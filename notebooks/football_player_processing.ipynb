```python
import pandas as pd
from unidecode import unidecode

# CSV 파일 불러오기
df = pd.read_csv('../data/fifa25_players_overall.csv', encoding='utf-8-sig')

# name 열에 대해 unidecode 적용
df['name'] = df['name'].apply(unidecode)

# 결과 확인 (선택사항)
print(df[['name']].head())

# 다시 저장
df.to_csv('../data/fifa25_players_overall_ascii.csv', index=False, encoding='utf-8-sig')
```
